# Plot

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
x = np.linspace(0, 5, 11)
y = x ** 2
```

### Vẽ các điểm và đường trong mặt phẳng(2D)
Phần lớn các câu lệnh để vẽ đồ thị trong mặt phẳng đều là lệnh plot. Lệnh plot vẽ đồ thị của một mảng dữ liệu trong một hệ trục thích hợp và nối các điểm bằng đường thẳng.

```python
x = np.linspace(0,2 * np.pi, 30) # Hàm linspace là lấy 30 điểm cách đều nhau trong khoảng từ 0 -> 2pi
y = np.sin(x)
plt.plot(x, y)

# Đồ thị nhận được được hiển thi trên cửa sổ figure:
```

***Lưu ý:*** Đối tượng 'Line2D' không có thuộc tính 'dpi'

**Chúng ta cũng có thể vẽ nhiều đồ thị bằng cách thêm:**

```python
z = np.cos(x)
plt.plot(x, y, x, z)
```

**Nếu thay đổi trật tự các đối số đồ thị sẽ xoay một góc 90**

```python
plt.plot(y,x,z,x)
```

MATLAB mặc định đường vẽ là đường liền, không đánh dấu, màu xanh da trời.Ta có thể thay đổi kiểu đường vẽ và đánh dấu lên đồ thị bằng cách đưa vào một đối số thứ ba. Các đối số tùy chọn này là một xâu kí tự, có thể chứa một hoặc nhiều hơn theo bảng dưới đây.

```python
plt.plot(x,y,'m*',x,y,'b--')
```

### Plot range
We can configure the ranges of the axes using the `set_ylim` and `set_xlim` methods in the axis object, or `axis('tight')` for automatically getting "tightly fitted" axes ranges:

```python
fig, axes = plt.subplots(1, 3, figsize=(12, 4))

axes[0].plot(x, x**2, x, x**3)
axes[0].set_title("default axes ranges")

axes[1].plot(x, x**2, x, x**3)
axes[1].axis('tight')
axes[1].set_title("tight axes")

axes[2].plot(x, x**2, x, x**3)
axes[2].set_ylim([0, 60])
axes[2].set_xlim([2, 5])
axes[2].set_title("custom axes range");
```

### Special Plot Types

There are many specialized plots we can create, such as barplots, histograms, scatter plots, and much more. Most of these type of plots we will actually create using seaborn, a statistical plotting library for Python. But here are a few examples of these type of plots:

```python
plt.scatter(x, y)
```

```python
from random import sample
data = sample(range(1, 1000), 100)
plt.hist(data)
```

```python
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# rectangular box plot
plt.boxplot(data,vert=True,patch_artist=True);   
```
