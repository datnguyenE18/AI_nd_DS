```python
import pandas as pd
import numpy as np
from numpy.random import randn
```

```python
df = pd.DataFrame(randn(5, 4), index = 'A B C D E'.split(), columns = 'W X Y Z'.split())
# or: df = pd.DataFrame(randn(5, 4), ['A', 'B', 'C', 'D', 'E'], ['W', 'X', 'Y', 'Z'])

# Show: 
df
```

## reset_index()

*Inplace:*

```python
df.reset_index()
```

```python
df # not changed
```

```python
temp = df.reset_index()
# Show:
temp
```

```python
# or:
df.reset_index(inplace = True)
df # be changed
```

*Drop:*

```python
# Creat new df:
df = pd.DataFrame(randn(5, 4), 'A B C D E'.split(), 'W X Y Z'.split())
df
```

```python
df.reset_index()
```

```python
df.reset_index(drop = True)
```
