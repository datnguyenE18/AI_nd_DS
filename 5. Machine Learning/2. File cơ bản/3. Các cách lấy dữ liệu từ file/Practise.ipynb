### C1: Sử dụng csv: 

```python
X = []
Y = []
with open('ex2data1.csv', mode='r') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    for row in csv_reader:
        X.append([float(row['x1']), float(row['x2'])])
        Y.append(float(row['y']))
type(X), X, type(X), Y
```

```python
X = np.array(X)
Y = np.array(Y)
#print(type(X)), print(type(Y)) 
type(X), X, type(X), Y
```
