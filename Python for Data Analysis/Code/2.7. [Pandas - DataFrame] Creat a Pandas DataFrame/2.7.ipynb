*In the real world, a Pandas DataFrame will be created by loading the datasets from existing storage, storage can be SQL Database, CSV file, and Excel file. Pandas DataFrame can be created from the lists, dictionary, and from a list of dictionary etc. Dataframe can be created in different ways*

```python
import pandas as pd
import numpy as np
```

**General way:**

```python
# Import randn from random (Numpy)
from numpy.random import randn

df = pd.DataFrame(100 + 10, index = 'A B C D E'.split(), columns = 'W X Y Z'.split())
#or: df = pd.DataFrame(100, index = ['A', 'B', 'C', 'D', 'E'], columns = ['W', 'X', 'Y', 'Z'])

# Show: 
df
```

```python
pd.DataFrame(randn(5, 4), index = [2, 8, 9, 10, 20], columns = 'W X Y Z'.split())
```

```python
pd.DataFrame(randn(5, 6), index = [1, 2, 3, 4, 5], columns =[1, 2, 3, 4, 5, 6])
```

```python
np.random.seed(101)
pd.DataFrame(randn(3, 2), index = [1, 2, 3], columns =[1, 2])
```

```python
np.random.seed(101)
print (randn(3, 2))
```

**Using List:**

```python
lst = [['VietNam', 'America', 'England', 'Franch', 'China'], ['Korea', 'Japan', 'Italy', 'Germany', 'Iceland']]
```

```python
lst
```

```python
df = pd.DataFrame( data = lst, columns = 'A B C D E'.split())
df
```

**From dict of ndarray/lists:**

```python
data = {
    'Name' : 'Tuan Anh Dam Duc An'.split(), 
    'Age' : [20, 21, 23, 25, 26],
    'Score' : [100, 99, 95, 93, 97]
}
```

```python
df = pd.DataFrame(data, index = '1st 2nd 3rd 4th 5th'.split())
df
```
