### **Tính đạo hàm của log(x)**

```python
x = tf.constant(-3.) # Định nghĩa 1 biến trong tf
x
```

```python
with tf.GradientTape() as g:
  g.watch(x) # đạo hàm trên x hay trên y
  f = tf.math.log(x) # Định nghĩa hàm
df_dx = g.gradient(f, x) # tham số f là kết quả, tham số x là đầu vào
```

```python
df_dx
```

### **Đạo hàm của 1 / x**

```python
x = tf.constant(4.)
with tf.GradientTape() as g:
  g.watch(x)
  f = 1/x
df_dx = g.gradient(f, x) # = -1 / x^2
df_dx # = -1 / 16
```

```python
# cần lưu ý, đạo tại 0:

x = tf.constant(0.)
with tf.GradientTape() as g:
  g.watch(x)
  f = tf.math.abs(x) # |x|
df_dx = g.gradient(f, x) # = x / |x|
df_dx
```

```python
x = tf.constant(0.)
with tf.GradientTape() as g:
  g.watch(x)
  f = 1/x
df_dx = g.gradient(f, x) # = -1 / x^2
df_dx # nó sẽ không báo lỗi mà ra kết quả là -Inf
```

```python
x = tf.constant(1.0)
with tf.GradientTape() as g:
  g.watch(x)
  g1 = 1 / x
  g2 = tf.math.log(g1)

dg2_dx = g.gradient(g2, x)
dg2_dx
```

### **Đạo hàm riêng:**

1.png

```python
x1 = tf.constant(1.)
x2 = tf.constant(1.)
with tf.GradientTape() as g:
  g.watch(x1)
  f = 3 * (x1**3) * (x2**3) - (x1**0.5) * (x2**2)

df_dx = g.gradient(f, x1)
df_dx
```
