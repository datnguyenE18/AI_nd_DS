# subplot( ) & subplots( )

**subplot ( )** 

```python
import matplotlib.pyplot as plt
import numpy as np
```

Chúng ta có thể sử dụng phương thức **subplot( )** để thêm nhiều **plots* trong một hình.

Phương thức **subplot( )** có ba đối số: **nrows**, **ncols** và **index**. Chúng chỉ ra số lượng hàng, số cột và số index của sub-plot.

```python
plt.subplot(2, 1, 1) # 2 hàng, 1 cột và ở vị trí đầu tiên (trên đầu)
plt.subplot(2, 1, 2) # 2 hàng, 1 cột và ở vị trí thứ hai (bên dưới)
```

**Ví dụ:** ta muốn tạo hai sub-plot trong một hình sao cho nó nằm trên một hàng và trên hai cột và do đó ta chuyển các đối số (1,2,1) và (1,2,2) trong phương thức **subplot( )**. Lưu ý rằng ta đã sử dụng riêng phương thức **title( )** cho cả các **subplots**. Ta sử dụng phương thức **suptitle( )** để tạo một tiêu đề chung cho hình.

```python
x = np.arange(1, 5)
y = x**3
```

```python
plt.subplot(1, 2, 1)
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'go')
plt.title('1st subplot')

plt.subplot(1, 2, 2)
plt.plot(x, y, 'r^')
plt.title('2nd subplot')

plt.suptitle('My sub_plots')
# plt.show()
```

**subplots( )**

Khi chúng ta muốn có nhiều **subplots** trong hình. Một cách thuận tiện hơn là sử dụng phương thức **subplots()**. 

Lưu ý về sự khác biệt của các giá trị trong cả hai phương thức: Phương thức này lấy hai đối số **nrows** và **ncols** làm số lượng hàng và số cột tương ứng. Phương thức này tạo ra hai đối tượng: **figure** và **axes** mà chúng ta lưu trữ trong các biến **fig** và **ax** có thể được sử dụng để thay đổi các thuộc tính mức **figure** và **axes** tương ứng. Lưu ý rằng các tên biến này được chọn tùy ý.

```python
fig, axes = plt.subplots()
```

```python
# Basic:
fig, axes = plt.subplots()

axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title');
```

```python
# Axes is an array of axes to plot on
axes
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 2, figsize = (8, 6), dpi = 120)
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 2, figsize = (8, 6), dpi = 120)

axes[0, 1].plot([1, 2, 3, 4], [1, 4, 9, 16], '-go')
axes[1, 0].plot(x, y, 'r^')
axes[0, 1].set_title('Squares')
axes[1, 0].set_title('Cubes')
#plt.show()
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2)

# Sử dụng vòng lặp:
for ax in axes:
    ax.plot(x, y, 'b')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_title('title')
fig
```

Có một vấn đề phổ biến với **matplolib** là các ô con hoặc số liệu chồng chéo. Vì vậy chúng ta có thể sử dụng phương thức **fig.tight_layout ()** hoặc **plt.tight_layout ()**, tự động điều chỉnh vị trí của các trục trên khung vẽ hình để không có nội dung chồng chéo:

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2)

# Sử dụng vòng lặp:
for ax in axes:
    ax.plot(x, y, 'b')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_title('title')
fig 
fig.tight_layout()
```
