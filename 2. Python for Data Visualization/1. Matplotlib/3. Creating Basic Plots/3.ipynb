### Basic Example

Let's walk through a very simple example using two numpy arrays. You can also use lists, but most likely you'll be passing numpy arrays or pandas columns (which essentially also behave like arrays).

```python
plt.plot( [1, 2, 3, 6, 10], [1, 4, 7, 8, 8])
```

Một vài tham số về định dạng:

- tham số màu :  **r - red**, **g -green**, **b - blue**, **y - yellow**, …

- định dạng đường :  **- (đường thẳng)**, **-- (đường nét đứt)**, nếu không truyền tham số là không vẽ đường.

- định dạng đỉnh :  * **(đỉnh hình bông tuyết)**, **D (đỉnh hình thoi)**, **s (đỉnh hình vuông)**, **go (chấm tròn)**, ...

```python
plt.plot( [1, 2, 3, 6, 10], [1, 4, 7, 8, 8], 'r')
```

```python
plt.plot( [1, 2, 3, 6, 10], [1, 4, 7, 8, 8], '--')
```

```python
plt.plot( [1, 2, 3, 6, 10], [1, 4, 7, 8, 8], '-D')
```

```python
plt.plot( [1, 2, 3, 6, 10], [1, 4, 7, 8, 8], '--*')
```

```python
plt.plot( [1, 2, 3, 6, 10], [1, 4, 7, 8, 8], '--sr')
```

## Basic Matplotlib Commands

**The data we want to plot:**

```python
import numpy as np
x = np.linspace(0, 5, 11)
y = x ** 3 + 1
```

```python
x
```

```python
y
```

```python
plt.plot(x, y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Basic Example')
# plt.show() 
```

```python
plt.plot(x, y, 'r*--')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Basic Example')
plt.show()  # it like you are Printing a plot
```

### Creating Multiplots on Same Canvas

``plt.subplot(nrows, ncols, plot_number)``

```python
plt.subplot(3,4,1)
```

```python
plt.subplot(1,2,1)
plt.plot(x, y, 'r--') # More on color options later
plt.subplot(1,2,2)
plt.plot(y, x, 'g*-');
```

### Creating Multi-Objects on Same Plot

```python
plt.plot([1, 2, 3, 4, 5], [6, 10, 2, 1, 10], 'go-', label = 'GreenDots')
plt.plot([3, 8, 6], [11, 9, 20], 'b*-', label = 'Bluestars')
plt.plot([2, 6, 5, 7, 9], [11, 2, 6, 9, 1], 'y*-', label = 'GreenDots')
plt.title('Example')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend(loc = 'best') # thêm chú thích
# plt.show()
```

- tham số label trong hàm **plt.plot()** là để đặt tên cho một điểm, tên này sẽ xuất hiện trong chú thích.

- **plt.xlabel** hay **plt.ylabel** là để đặt tên cho trục x và trục y.

- **plt.title** là để đặt tên cho biểu đồ

- **plt.legend** sẽ đi tìm các thành phần chứa tham số label và đưa vào chú thích
