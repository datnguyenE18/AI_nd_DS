# Basic Image Processing

```python
import cv2
```

### Read an Image

```python
# Đối tượng 'img' chứa thông tin ảnh
img = cv2.imread('1.jpg')
```

### Display an Image

***Use CV2***

```python
# cv2.imshow(window_name, image)
cv2.imshow('Beautiful', img)
cv2.waitKey(0)
```

*Hàm waitKey(0) là hàm chờ không cho thoát cửa sổ lập tức mà phải người dùng nhấn phím bất kỳ để thoát.*

**Use matplotlib**

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(img)
```

*CV2 mặc định đọc ảnh màu theo thứ là BGR (là blue, green và red) nên ảnh có màu hơi xanh.*

Vì vậy ta phải viết thế này:

```python
plt.imshow(img[:,:,::-1])
```
