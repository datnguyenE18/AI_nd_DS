# Chú thích trực tiếp trong hình

Việc thêm những dòng chữ chú thích vào trực tiếp trong hình sẽ giúp biểu đồ trở nên rõ ràng và dễ hiểu hơn bằng cách sử dụng **plt.text()** 

Hai tham số chính là **tọa độ** và **nội dung đoạn text**, ngoài ra có thể thêm vào các tham số như màu sắc hay căn lề. 

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
import numpy as np
```

```python
plt.figure(figsize = (7, 4), dpi = 120)
X = np.linspace(0, 8*np.pi, 1000) 
sine = plt.plot(X, np.sin(X), color = 'blue');
                                                                                
for angle in [440, 810, 1170]:
    plt.text(angle/57.2985, 1.05, str(angle) + "\ndegrees", 
             transform = plt.gca().transData, horizontalalignment = 'center', color='green')
    
for angle in [270, 630, 990, 1350]:
    plt.text(angle/57.2985, -1.3, str(angle) + "\ndegrees",
             transform = plt.gca().transData, horizontalalignment = 'center', color='red')
                                                                                
plt.gca().set( ylim = (-2.0, 2.0), xlim = (-.5, 26))                                 
plt.title('add texts into a figure', fontsize = 18)                              
plt.show()
```
