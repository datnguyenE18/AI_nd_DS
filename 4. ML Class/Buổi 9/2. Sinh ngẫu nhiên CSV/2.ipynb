```python
import random
import csv
```

```python
# B1: Tạo / mở file để ghi.
with open('randomPatient.csv', mode='w', newline='') as patient_file:
    patient_writer = csv.writer(patient_file)
    
    # B2: Tạo cái tên cột:
    patient_writer.writerow(['ID',
                             'Nam sinh',
                             'Chieu cao',
                             'Can nang',
                             'Nhiet do'])
    
    
    n = 10 # số đối tượng (người)
    
    # B3: random từng cột từng người một trong 10 người:
    for i in range(n):
        patient_writer.writerow([i, 
                                 random.randint(1930, 2000),
                                 random.randint(130, 220),
                                 random.randint(30, 120),
                                 random.uniform(36.5, 43.1)])
```
