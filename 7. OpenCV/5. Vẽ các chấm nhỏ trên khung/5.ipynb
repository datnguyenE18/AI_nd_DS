```python
import cv2
import numpy as np
import random
```

```python
d = 500 # Độ lớn của hình vuông
shw = np.zeros(shape = (d, d, 3))

for i in range(d):
    for j in range(d):
        shw[i][j] = 255
```

**Vẽ các chấm tròn bên trong khung hình vuông:**

```python
N = 20 # Số chấm tròn sẽ sinh

for i in range(N):
    # sinh các tung độ và hoành độ của điểm nắm trong khung hình vuông một cách ngẫu nhiên
    rand_x = random.randint(0, 500)
    rand_y = random.randint(0, 500)
    
    cv2.circle(shw, (rand_x, rand_y), 2,  # sinh ra những hình tròn kích thước rất nhỏ (coi là điểm)
               (255, 0, 0), -1) # màu xanh nước biển
    
    cv2.imshow("img", shw)
    cv2.waitKey(100)
```

```python
for i in range(N):
    # sinh các tung độ và hoành độ của điểm nắm trong khung hình vuông một cách ngẫu nhiên
    rand_x = random.randint(0, 500)
    rand_y = random.randint(0, 500)
    
    cv2.circle(shw, (rand_x, rand_y), 2, (0, 255, 0), -1) # màu xanh lá cây
    
    cv2.imshow("img", shw)
    cv2.waitKey(100)
```
