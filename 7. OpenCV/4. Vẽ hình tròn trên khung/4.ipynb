```python
import cv2
import numpy as np
```

```python
shw = np.zeros(shape = (500, 500, 3))
cv2.imshow("Test!", shw)
cv2.waitKey(0)
```

**Độ lớn và vị trí hình tròn:**

```python
R = 100 # Bán kính hình tròn 
x = 250 # Tung độ hình tròn
y = 350 # Toành độ hình tròn
```

**Hình trắng nền đen:**

```python
cv2.circle(shw, (x, y), R, 
           (255, 255, 255)) # màu của hình tròn

cv2.imshow("Circle", shw)
cv2.waitKey(0)
```

**Hình đen nền trắng:**

```python
# Chuyển nền sang màu trắng:
for i in range(500):
    for j in range(500):
        shw[i][j] = 255

# Vẽ hình tròn màu đen:
cv2.circle(shw, (x, y), R, (0, 0, 0)) 

cv2.imshow("Circle", shw)
cv2.waitKey(0)
```
