```python
import cv2
import numpy as np
```

**Kích thước hình vuông / HCN cần tạo:**

```python
rows = 500
cols = 400
```

**Tạo ma trận mô hình:**

```python
test1 = np.zeros(shape = (5, 10, 3))
test1
```

```python
test2 = np.zeros(shape = (rows, cols, 3))
test2
```

```python
test3 = np.zeros(shape = (rows, cols, 2))
test3
```

```python
test4 = np.zeros(shape = (rows, cols, 3), dtype = int)
test4
```

```python
test5 = np.zeros(shape = (rows, cols, 3), dtype = np.uint8)
test5
```

**In ra hình ảnh dựa vào mô hình trên:**

```python
cv2.imshow("img", test1)
cv2.waitKey(0)
```

```python
cv2.imshow("img", test2)
cv2.waitKey(0)
```
